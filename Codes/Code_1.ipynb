```python
import pandas as pd
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots
```

```python
df = pd.read_csv('Permanent_Crop_NUTS2.csv', sep=",")
df.head(50)


```

```python
# Filtrar los datos por los aos deseados
years = [1989, 1999, 2009, 2019]
df_filtered = df[df['year'].isin(years)]

# Configurar el tamao de la figura y crear subplots
plt.figure(figsize=(16, 10))

for i, year in enumerate(years, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='region_name', y='area', data=df_filtered[df_filtered['year'] == year])
    plt.title(f'Ao {year}')
    plt.xlabel('Regin')
    plt.ylabel('rea sembrada')

plt.tight_layout()
plt.show()
```
